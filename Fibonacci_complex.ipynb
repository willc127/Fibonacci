```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fibonacci(n):
    fib = np.zeros(n)
    fib[0] = 0
    fib[1] = 1

    for i in range(2, n):
        fib[i] = fib[i-1] + fib[i-2]

    fib = np.array(fib, dtype=int)
    return fib
```

```python
def binet_fibonacci(n):
    # Calcula o valor de Fibonacci para um número real n usando a fórmula de Binet
    phi = (1 + np.sqrt(5)) / 2  # Proporção áurea
    psi = (1 - np.sqrt(5)) / 2  # Complemento da proporção áurea
    
    # Usa np.emath.power para lidar com potências fracionárias de números negativos
    phi_n = np.emath.power(phi, n)
    psi_n = np.emath.power(psi, n)
    
    # Calcula a fórmula de Binet
    return (phi_n - psi_n) / np.sqrt(5)
```

```python
# Intervalo de valores reais para n
n_values = np.linspace(0, 6, 1000)  # De -5 a 5, com 1000 pontos

# Calcula F(n) para cada valor de n
fib_values = binet_fibonacci(n_values)
fib = fibonacci(int(max(n_values)+1))

# Separa as partes real e imaginária
real_part = np.real(fib_values)
imag_part = np.imag(fib_values)
```

```python
# Plot da parte real e imaginária

plt.figure(figsize=(13, 5))
plt.plot(real_part, imag_part, color = 'black')
plt.hlines(0, real_part[0], real_part[-1], color='black',linestyles='dashed' , linewidth=1)
plt.plot(fib,fib*0,'or', label='Fibonacci sequence')
plt.xlim(real_part[0]-0.05,real_part[-1]+0.05)
plt.xlabel("Re(F(n))")
plt.ylabel("Im(F(n))")
```
